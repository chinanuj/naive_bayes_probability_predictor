```python
# Writing a generalized function to calculate posterior probability by taking reference from  above manual calculations
def calculate_posterior_probabilities(df, feature_columns, class_column, input_combination):
    # Count occurrences of features for the given input combination
    counts = {}
    for feature_column in feature_columns + [class_column]:
        counts[feature_column] = {}
        for feature_value in df[feature_column].unique():
            counts[feature_column][feature_value] = df[(df[feature_column] == feature_value)].shape[0]

    # Calculate individual probabilities for each feature value
    individual_probabilities = {}
    for feature_column, feature_values in counts.items():
        individual_probabilities[feature_column] = {feature_value: count / len(df) for feature_value, count in feature_values.items()}

    # Calculate likelihood probabilities for the given input combination
    likelihood_probabilities = {}
    for class_value in df[class_column].unique():
        likelihood_probabilities[class_value] = 1
        for feature_column, feature_value in input_combination.items():
            count_feature_and_class = df[(df[feature_column] == feature_value) & (df[class_column] == class_value)].shape[0]
            count_class = df[df[class_column] == class_value].shape[0]
            likelihood_probabilities[class_value] *= count_feature_and_class / count_class

    # Calculate prior probabilities
    prior_probabilities = {}
    for class_value in df[class_column].unique():
        prior_probabilities[class_value] = counts[class_column][class_value] / len(df)

    # Calculate posterior probabilities
    posterior_probabilities = {}
    for class_value in df[class_column].unique():
        posterior_probabilities[class_value] = (likelihood_probabilities[class_value] * prior_probabilities[class_value]) / \
                                               (individual_probabilities['Outlook'][input_combination['Outlook']] *
                                                individual_probabilities['Temp'][input_combination['Temp']] *
                                                individual_probabilities['Humidity'][input_combination['Humidity']] *
                                                individual_probabilities['Windy'][input_combination['Windy']])

    return posterior_probabilities

# Example usage:
df = pd.DataFrame({
    'Outlook': ['Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Sunny', 'Rainy', 'Overcast', 'Overcast', 'Sunny'],
    'Temp': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Windy': ['f', 't', 'f', 'f', 'f', 't', 't', 'f', 'f', 'f', 't', 't', 'f', 't'],
    'Play': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
})

input_combination1 = {'Outlook': 'Overcast', 'Temp': 'Mild', 'Humidity': 'High', 'Windy': 't'}
input_combination2 = {'Outlook': 'Sunny', 'Temp': 'Mild', 'Humidity': 'Normal', 'Windy': 'f'}

posterior_probs1 = calculate_posterior_probabilities(df, ['Outlook', 'Temp', 'Humidity', 'Windy'], 'Play', input_combination1)
posterior_probs2 = calculate_posterior_probabilities(df, ['Outlook', 'Temp', 'Humidity', 'Windy'], 'Play', input_combination2)

# Print posterior probabilities
for class_value, posterior_prob in posterior_probs1.items():
    print(f"Posterior probability of {class_value} given Outlook={input_combination1['Outlook']}, Temp={input_combination1['Temp']}, Humidity={input_combination1['Humidity']}, Windy={input_combination1['Windy']}: {posterior_prob}")

# Print posterior probabilities
for class_value, posterior_prob in posterior_probs2.items():
    print(f"Posterior probability of {class_value} given Outlook={input_combination2['Outlook']}, Temp={input_combination2['Temp']}, Humidity={input_combination2['Humidity']}, Windy={input_combination2['Windy']}: {posterior_prob}")
```

```python
# Writing a generalized function to calculate posterior probability with laplace smoothing by taking reference from  above calculations

def calculate_likelihood_probabilities_laplace(df, feature_columns, class_column, alpha=1):
    likelihood_probabilities = {}

    for feature_column in feature_columns:
        unique_features = df[feature_column].unique()
        total_unique_features = len(unique_features)  # Total unique features for Laplace smoothing
        for feature in unique_features:
            for class_value in df[class_column].unique():
                count_class_and_feature = df[(df[feature_column] == feature) & (df[class_column] == class_value)].shape[0]
                count_class = df[df[class_column] == class_value].shape[0]

                # Calculate likelihood probability with Laplace Smoothing
                likelihood_probability = (count_class_and_feature + alpha) / (count_class + alpha * total_unique_features)

                # Create key in the format "probability of feature_value and class_value"
                key = f"probability of {feature_column} = {feature} and {class_column} = {class_value}"
                likelihood_probabilities[key] = likelihood_probability

    return likelihood_probabilities


# Example usage
feature_columns = ['Outlook', 'Temp', 'Humidity', 'Windy']
likelihood_probabilities = calculate_likelihood_probabilities_laplace(df, feature_columns, 'Play')
for key, value in likelihood_probabilities.items():
    print(f"{key} : {value}")


def calculate_posterior_probabilities(df, feature_columns, class_column, input_combination, alpha=1):
    # Calculate likelihood probabilities with Laplace Smoothing
    likelihood_probabilities = calculate_likelihood_probabilities_laplace(df, feature_columns, class_column, alpha)

    # Calculate prior probabilities
    prior_probabilities = {}
    total_samples = len(df)
    for class_value in df[class_column].unique():
        count_class = df[df[class_column] == class_value].shape[0]
        prior_probabilities[class_value] = count_class / total_samples

    # Calculate posterior probabilities
    posterior_probabilities = {}
    for class_value in df[class_column].unique():
        posterior_probabilities[class_value] = 1
        for feature_column, feature_value in input_combination.items():
            key = f"probability of {feature_column} = {feature_value} and {class_column} = {class_value}"
            posterior_probabilities[class_value] *= likelihood_probabilities[key]

        # Multiply by the prior probability
        posterior_probabilities[class_value] *= prior_probabilities[class_value]

    # Normalize probabilities
    sum_probabilities = sum(posterior_probabilities.values())
    for class_value in posterior_probabilities:
        posterior_probabilities[class_value] /= sum_probabilities

    return posterior_probabilities

# Example usage:
df = pd.DataFrame({
    'Outlook': ['Rainy', 'Rainy', 'Overcast', 'Sunny', 'Sunny', 'Sunny', 'Overcast', 'Rainy', 'Rainy', 'Sunny', 'Rainy', 'Overcast', 'Overcast', 'Sunny'],
    'Temp': ['Hot', 'Hot', 'Hot', 'Mild', 'Cool', 'Cool', 'Cool', 'Mild', 'Cool', 'Mild', 'Mild', 'Mild', 'Hot', 'Mild'],
    'Humidity': ['High', 'High', 'High', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'Normal', 'Normal', 'High', 'Normal', 'High'],
    'Windy': ['f', 't', 'f', 'f', 'f', 't', 't', 'f', 'f', 'f', 't', 't', 'f', 't'],
    'Play': ['no', 'no', 'yes', 'yes', 'yes', 'no', 'yes', 'no', 'yes', 'yes', 'yes', 'yes', 'yes', 'no']
})

input_combination = {'Outlook': 'Overcast', 'Temp': 'Mild', 'Humidity': 'High', 'Windy': 't'}

posterior_probs = calculate_posterior_probabilities(df, ['Outlook', 'Temp', 'Humidity', 'Windy'], 'Play', input_combination)

# Print posterior probabilities
for class_value, posterior_prob in posterior_probs.items():
    print(f"Posterior probability of {class_value} given input combination {input_combination}: {posterior_prob}")
```
